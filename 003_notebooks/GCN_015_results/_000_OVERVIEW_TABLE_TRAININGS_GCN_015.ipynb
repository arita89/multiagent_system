## only model information

```python
mask = (df.model == GCN_MODEL) #and (df.Nepochs == 5000)
df_model = df[mask]
df_model
```

```python
## min loss for each criterion:
dict_best_runs_per_criterion = {}
#list(df_model.criterion.unique())
for criterion in sorted(["MSE","L1","Custom_Loss_6"]):
    print (DASH)
    print (criterion)
    
    mask = df_model.criterion == criterion
    df_model_criterion = df_model[mask]
    
    # check the min in training 
    abs_min_train = df_model_criterion[columns_with_final_train_losses].min().min() 
    for col in columns_with_final_train_losses:
        try: 
            min_col = df_model_criterion[col].min()
            if min_col == abs_min_train:
                search_col = col
        except Exception as e:
            print ("")
            #print (e)
    index_of_min_train_value = df_model_criterion[df_model_criterion[search_col] == abs_min_train].index.tolist()[0]
    best_model_train = df_model_criterion.loc[index_of_min_train_value].to_dict() #["final_train_loss"]
    print (f'{best_model_train["run_date"]=}')
    print (f'{best_model_train["run_time"]=}')
    print (f'{best_model_train["reduction"]=}')
    print (f"{abs_min_train=}")
    
    # check the min in validation 
    abs_min_val = df_model_criterion[columns_with_final_val_losses].min().min()
    # in which column it is found
    counter = 0
    for col in columns_with_final_val_losses:
        try: 
            min_col = df_model_criterion[col].min()
            if min_col == abs_min_val:
                search_col = col
        except Exception as e:
            counter +=1
            #print ("")
    index_of_min_val_value = df_model_criterion[df_model_criterion["final_val_loss"] == abs_min_val].index.tolist()[0]
    best_model_val = df_model_criterion.loc[index_of_min_val_value].to_dict()
    print (f'{best_model_val["run_date"]=}')
    print (f'{best_model_val["run_time"]=}')
    print (f'{best_model_val["reduction"]=}') 
    print (f"{abs_min_val=}")

    
    # append the information 
    dict_best_runs_per_criterion[criterion] = (index_of_min_train_value, index_of_min_val_value)
```
